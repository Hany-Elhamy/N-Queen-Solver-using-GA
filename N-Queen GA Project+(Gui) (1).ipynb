```python
from random import randint
from random import random
def create_solution():
    solution=[]
    while (len(solution)!=n):
        z=randint(1,n)
        if z not in solution:
            solution.append(z)
            
    return(solution)


'''x=create_solution()
x'''
```

```python
def population(count):
    return[create_solution() for i in range(count)]
#population(10)
```

```python
def plot_sol(solution):
    board= [[0] * n for _ in range(n)]
    
    for i in range(len(solution)):
        board[solution[i]-1][i]=1
    return board
```

```python
def fitness(solution):
    board=plot_sol(solution)
    penalty=0
    for z in range(len(solution)):
        row=solution[z]-1
        col=z
        for i in range(col):
            if board[row][i] == 1:
                penalty+=1
                break
        for i, j in zip(range(row-1, -1, -1),range(col-1, -1, -1)):
            if board[i][j] == 1:
                penalty+=1
                break

        for i, j in zip(range(row+1, n, 1),range(col-1, -1, -1)):
            if board[i][j] == 1:
                penalty+=1
                break

    return (penalty)
```

```python
def avg_fitness(population):
    all_solutions=[fitness(solution) for solution in population]
    avg_of_fit=sum(all_solutions)/len(all_solutions)*1.0
    return(avg_of_fit)
    
```

```python
def evolve(population,retain=0.1,mutation=0.01,random_selection=0.05):
    
    #top no. of parents
    gradedpop=[(fitness(solution),solution) for solution in population ]
    gradedpop=[x[1] for x in sorted(gradedpop)]
    retain_no=int(len(gradedpop)*retain)
    parents=gradedpop[:retain_no]
    
    #random selection for diversity
    for sol in gradedpop[retain_no:]:
        if random_selection>random():
            parents.append(sol)
            
     
    # mutation
    for sol in parents:
        if mutation>random():
            num1=randint(0,n-1)
            num2=randint(0,n-1)
            sol[num1], sol[num2]=sol[num2],sol[num1]  
        
       # Crossover
    parents_length = len(parents)
    desired_length = len(population) - parents_length 
    children = []

    while len(children) < desired_length:
        male_index = randint(0, parents_length - 1)
        female_index = randint(0, parents_length - 1) 

        if male_index != female_index:
            male = parents[male_index] 
            female = parents[female_index] 
            crossover_point = randint(0, len(male) - 1)

            # Child 1
            child1 = male[:crossover_point]
            for i in female[crossover_point:]:
                if i not in child1:
                    child1.append(i)

            while len(child1) < len(male):
                for i in female[:crossover_point]:
                    if i not in child1:
                        child1.append(i)

            # Child 2
            child2 = female[:crossover_point]
            for i in male[crossover_point:]:
                if i not in child2:
                    child2.append(i)

            while len(child2) < len(female):
                for i in male[:crossover_point]:
                    if i not in child2:
                        child2.append(i)

            children.append(child1)
            children.append(child2)

    parents.extend(children)
    return parents

   
    
```

```python
def top1(pop):
    gradedpop=[(fitness(solution),solution) for solution in pop ]
    gradedpop=[x[1] for x in sorted(gradedpop)]
    top=[]
    for i in gradedpop:
        if len(top) >= 5:
            break
        elif i not in top:
            top.append(i)
            pprint(i)
     
    pprint(plot_sol(top[0]))   
    draw_board(plot_sol(top[0]))      
```

```python
entry = None
entry0= None
entry1= None
entry2= None
entry3= None
entry4= None
from pprint import pprint
import time
#for different 4-Queens create population with 50 members and better with mutation rate 0.01
#for different 16 Queens increase mutation rate to 0.09 
def run():
    
    global n
    n = int(entry.get())
    retain=float(entry1.get())
    mutation=float(entry2.get())
    random_selection=float(entry3.get())
    result_text.delete(1.0, tk.END)
    result_text.insert(tk.END, "Running...\n")
    root.update_idletasks()
    st = time.time()
    pop_count=int(entry0.get())
    num_of_iterations=int(entry4.get())
    pop=population(pop_count)
    fitness_history=[avg_fitness(pop)]
    for i in range(num_of_iterations):
        pop=evolve(pop,retain,mutation,random_selection)
        fitness_history.append(avg_fitness(pop))
    et = time.time()
    elapsed_time = et - st
    result_text.insert(tk.END, f'Execution time is: {elapsed_time:.4f} seconds\n\n')

    top1(pop)
    result_text.insert(tk.END, "\nFitness History:\n")
    for i, fitness_val in enumerate(fitness_history):
        result_text.insert(tk.END, f"Iteration {i}: {fitness_val}\n")
    
```

```python
def draw_board(board):
    canvas.delete("all")
    cell_size = 20
    board_size = len(board) * cell_size
    start_x = (400 - board_size) / 2
    start_y = (400 - board_size) / 2

    for row_index, row in enumerate(board):
        for col_index, value in enumerate(row):
            x0 = start_x + col_index * cell_size
            y0 = start_y + row_index * cell_size
            x1 = x0 + cell_size
            y1 = y0 + cell_size
            canvas.create_rectangle(x0, y0, x1, y1, fill="white" if (row_index + col_index) % 2 == 0 else "lightgray")
            if value == 1:
                canvas.create_oval(x0 + 5, y0 + 5, x1 - 5, y1 - 5, fill="red")


```
